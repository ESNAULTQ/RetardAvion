```python
# Sélectionner la cible (y)
y = df_cleaned3['ARR_DELAY_ENCODED']

# Sélectionner uniquement les caractéristiques numériques
X = df_cleaned3.select_dtypes(include=['float64', 'int64']).drop(columns=['ARR_DELAY_NEW', 'ARR_DELAY_ENCODED'])


# Combiner X et y pour faciliter l'undersampling
data = pd.concat([X, y], axis=1)

# Séparer les exemples de chaque classe
class_0 = data[data['ARR_DELAY_ENCODED'] == 0]
class_1 = data[data['ARR_DELAY_ENCODED'] == 1]

# Rééchantillonner la classe majoritaire (0)
class_0_undersampled = resample(class_0,
                                replace=False,  # échantillonnage sans remplacement
                                n_samples=len(class_1),  # égaliser le nombre d'exemples de la classe minoritaire
                                random_state=1)  # pour la reproductibilité

# Combiner les classes rééchantillonnées
undersampled_data = pd.concat([class_0_undersampled, class_1])

# Séparer les caractéristiques (X) et la cible (y) dans les données équilibrées
X_undersampled = undersampled_data.drop(columns=['ARR_DELAY_ENCODED'])
y_undersampled = undersampled_data['ARR_DELAY_ENCODED']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_undersampled, y_undersampled, test_size=0.3, random_state=1, stratify=y_undersampled)

# Initialiser et entraîner le modèle de régression logistique
logistic_regression_model = LogisticRegression(random_state=1, max_iter=1000)
logistic_regression_model.fit(X_train, y_train)

# Prédire les valeurs pour l'ensemble de test
y_pred = logistic_regression_model.predict(X_test)

# Évaluer le modèle
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Créer la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```

```python
# Sélectionner la cible (y)
y = df_cleaned3['ARR_DELAY_ENCODED']

# Sélectionner uniquement les caractéristiques numériques
X = df_cleaned3.select_dtypes(include=['float64', 'int64']).drop(columns=['ARR_DELAY_NEW', 'ARR_DELAY_ENCODED'])

# Combiner X et y pour faciliter l'oversampling
data = pd.concat([X, y], axis=1)

# Séparer les exemples de chaque classe
class_0 = data[data['ARR_DELAY_ENCODED'] == 0]
class_1 = data[data['ARR_DELAY_ENCODED'] == 1]

# Définir le nombre d'échantillons cible pour chaque classe
target_samples = max(len(class_0), len(class_1))

# Rééchantillonner la classe minoritaire (1) pour obtenir le même nombre d'exemples que la classe majoritaire (0)
class_1_oversampled = resample(class_1,
                               replace=True,  # échantillonnage avec remplacement
                               n_samples=target_samples - len(class_1),  # augmenter le nombre d'exemples de la classe minoritaire
                               random_state=1)  # pour la reproductibilité

# Combiner les classes rééchantillonnées
balanced_data = pd.concat([class_0, class_1, class_1_oversampled])

# Séparer les caractéristiques (X) et la cible (y) dans les données équilibrées
X_balanced = balanced_data.drop(columns=['ARR_DELAY_ENCODED'])
y_balanced = balanced_data['ARR_DELAY_ENCODED']

# Splitting the data
X_train, X_test, y_train, y_test = train_test_split(X_balanced, y_balanced, test_size=0.3, random_state=1, stratify=y_balanced)

# Initialiser et entraîner le modèle de régression logistique
logistic_regression_model = LogisticRegression(random_state=1, max_iter=1000)
logistic_regression_model.fit(X_train, y_train)

# Prédire les valeurs pour l'ensemble de test
y_pred = logistic_regression_model.predict(X_test)

# Évaluer le modèle
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))

# Créer la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)
```
